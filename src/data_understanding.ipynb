```python
import pandas as pd

# Load dataset
df = pd.read_csv("freMTPL2freq.csv")

# Basic inspection
df.head(), df.shape
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
df["ClaimOccurred"] = (df["ClaimNb"] > 0).astype(int)

df["ClaimOccurred"].value_counts(normalize=True)
```

```python
# Remove invalid exposure
df = df[df["Exposure"] > 0]

# Cap extreme values (common actuarial practice)
df["VehAge"] = df["VehAge"].clip(upper=20)
df["DrivAge"] = df["DrivAge"].clip(lower=18, upper=90)
df["BonusMalus"] = df["BonusMalus"].clip(upper=150)
```

```python
selected_cols = [
    "Exposure", "VehAge", "DrivAge", "BonusMalus", "Density",
    "Area", "VehBrand", "VehGas", "Region", "ClaimOccurred"
]

df[selected_cols].to_csv("underwriting_clean.csv", index=False)
```
